```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt   
import heapq
```

```python
def file2Matrix(path):
    data=pd.read_csv(path,sep='\t',header=None)
#     data.columns=['air','game','ice','like']
    return data.as_matrix()

data=file2Matrix('dating.txt')
# fig = plt.figure()
# ax = fig.add_subplot(111)
plt.scatter(data[:,0],data[:,1],5*np.array(data[:,-1]),np.array(data[:,-1]))
print(data.shape)
label=data[:,-1]
```

```python
def normalize(featureCol):
    Min=featureCol.min(0)
    Max=featureCol.max(0)
    normal=np.vectorize(lambda x: (x-Min)/(Max-Min))
    return normal(featureCol)
tmp=np.vstack((normalize(data[:,0]),normalize(data[:,1])))
data=np.transpose(np.vstack((normalize(data[:,2]),tmp)))
print(data.shape)
print(data[0,:])
plt.scatter(data[:,0],data[:,1],50*np.array(data[:,-1]),np.array(data[:,-1]))
```

```python
labeldPoints={}
for i in range(750):
    key=""
    for val in data[i,:]:
        key+=str(val)
        key+=','
    key=key.strip(',')
    labeldPoints[key]=label[i]
# print(labeldPoints)

class classifer:
    def __init__(self,labeldPoints,k):
        self.kN=k
        self.points=labeldPoints
    def lostfun(self,p1,p2):
        rank=len(p1)
        Sum=0
        for i in range (rank):
            Sum+=abs(float(p1[i])**rank-float(p2[i])**rank)
        return float(Sum**(1/rank))
    
    def predict(self,testPoint):
        losts=[]
        #point=[1,2,3]
        for point in self.points.keys():
            point=point.split(',')
            losts.append([self.lostfun(point,testPoint),point])
        losts.sort(key=lambda x: x[0])
        losts=losts[:self.kN]

        keys=[]
        for a in losts:
            key=''
            for val in a[1]:
                key+=str(val)
                key+=','
            key=key.strip(',')
            keys.append(key)
        result=[]
        for key in keys:
            result.append(labeldPoints[key])
        return max(result,key=result.count)
        
        
        
knn=classifer(labeldPoints,3)
knn.predict(['0.37973902845969665', '1.0', '1.0'])
cnt=0
predic=[]
for i in range(750,1000):
    pred=knn.predict(data[i,:])
    predic.append(pred)
    if pred==label[i]:
        cnt+=1
print(cnt/250)
plt.scatter(data[750:1000,0],data[750:1000,1],20*np.array(predic),np.array(predic))
```
